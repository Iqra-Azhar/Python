**Parsing website**

```python
url = 'https://www.imdb.com/search/title/?title_type=feature,tv_movie,short&release_date=START_DATE,END_DATE&adult=include&count=250&start=START_COUNT&ref_=adv_nxt'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}

response = requests.get(url,headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print(soup.prettify())
else:
   print(response.status_code)
```

**Movies Data between 2013-2023**

```python
movie_data = soup.find_all(class_='lister-item mode-advanced')
```

```python
# Movie Titles
movie_data = soup.find_all(class_='lister-item mode-advanced')
for movie_dataa in movie_data:
    title = movie_dataa.find('h3', class_='lister-item-header').text.strip()

    year_element = movie_dataa.find('span', class_='lister-item-year')
    if year_element:
        year_text = year_element.text.strip()

        # Movie Year
        year_start = year_text.find('20')  # Assuming all years start with '20'
        if year_start != -1:
            year = int(year_text[year_start:year_start + 4])

            # Year Range
            if 2013 <= year <= 2023:
                # Check if there is a 'strong' element with class 'rating'
                for movie_dataa in movie_data:
                 rating_element = movie_dataa.find('strong')
                 rating = rating_element.text.strip() if rating_element else "N/A"

                # Movie Genres
                genre_element = movie_dataa.find('span', class_='genre')
                genre = genre_element.text.strip() if genre_element else "N/A"

                # Movie Directors
                for movie_dataa in movie_data:
                 director_element = movie_dataa.find('p', class_='text-muted').find('a')
                 director = director_element.text.strip() if director_element else "N/A"


                # Displaying Movies Data
                print(f"Title: {title}")
                print(f"Year: {year}")
                print(f"Rating: {rating}")
                print(f"Genre: {genre}")
                print(f"Director: {director}\n")
```

**Data into csv**

```python
movie_data = []

# Movie Titles
movie_elements = soup.find_all(class_='lister-item mode-advanced')
for movie_dataa in movie_elements:
    title_element = movie_dataa.find('h3', class_='lister-item-header')
    title = title_element.a.text.strip() if title_element else "N/A"

    year_element = movie_dataa.find('span', class_='lister-item-year')
    if year_element:
        year_text = year_element.text.strip()

        # Movie Year
        year_start = year_text.find('20')  # Assuming all years start with '20'
        if year_start != -1:
            year = int(year_text[year_start:year_start + 4])

            # Year Range
            if 2013 <= year <= 2023:
                # Check if there is a 'strong' element with class 'rating'
                rating_element = movie_dataa.find('strong')
                rating = rating_element.text.strip() if rating_element else "N/A"

                # Movie Genres
                genre_element = movie_dataa.find('span', class_='genre')
                genre = genre_element.text.strip() if genre_element else "N/A"

                # Movie Directors
                director_element = movie_dataa.find('p', class_='').find('a')
                director = director_element.text.strip() if director_element else "N/A"

                # Append movie data to the list
                movie_data.append([title, year, rating, genre, director])
```

```python
df = pd.DataFrame(columns=['Movie Title', 'Release Year', 'IMDb Rating', 'Genre', 'Director'])

# Appending data into empty dataframe made earlier for scraped movies data
new_movies_df = pd.DataFrame(movie_data, columns=['Movie Title', 'Release Year', 'IMDb Rating', 'Genre', 'Director'])
df = df.append(new_movies_df, ignore_index=True)

# Save the updated DataFrame to a CSV file
df.to_csv('imdb_movies.csv', index=False)
```

**Average Rating**

```python
def clean_rating(rating):
    ratings = rating.split('.')  # Split ratings separated by periods
    ratings = [float(r.strip()) for r in ratings if r.strip().replace('.', '').isdigit()]  # Convert to float
    if ratings:
        return sum(ratings) / len(ratings)  # Calculate average
    else:
        return None

df['IMDb Rating'] = df['IMDb Rating'].apply(clean_rating)

# Calculate the average IMDb rating
average_rating = df['IMDb Rating'].mean()

# Display the average rating
print(f"Average IMDb Rating: {average_rating:.2f}")
```

**Common Genre**

```python
common_genre = df['Genre'].mode().iloc[0]
print(f"Most Common Genre: {common_genre}")
```

**Dir_Avg_Rating**

```python
director_avg_rating = df.groupby('Director')['IMDb Rating'].mean().idxmax()
print(f"Director with Highest Average Rating: {director_avg_rating}")
```

**Most watched moives year**

```python
year_most_movies = df['Release Year'].value_counts().idxmax()
print(f"Year with Highest Number of Top-Rated Movies: {year_most_movies}")
```
